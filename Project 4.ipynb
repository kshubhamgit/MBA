```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel data into a pandas dataframe
df = pd.read_excel('Survey Response.xlsx',sheet_name=2)
df.head()
```

```python
len(df)
```

```python
df.drop(columns = ["Timestamp"],axis=1,inplace=True)
```

```python
objList = df.select_dtypes(include = "object").columns
classes = {}

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    df[feat] = le.fit_transform(df[feat].astype(str))
    classes[feat] = le.classes_

print (df.info())
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

i=1

# Calculate the frequency of each value in the list
age_group_counts = pd.Series(df.iloc[:,i]).value_counts()

# Create a pie chart
plt.pie(age_group_counts.values, labels=classes[df.columns[i]], autopct='%1.1f%%')

# Add a title to the chart
plt.title('Distribution of Age Group')

# Display the chart
plt.show()
```
